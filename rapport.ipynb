## Youtube

YouTube est un site web d’hébergement de vidéos et un média social sur lequel les utilisateurs peuvent envoyer, regarder, commenter, évaluer et partager des vidéos. 

### Chiffres clés

* créé aux États-Unis en 2005
* racheté par Google en octobre 2006
* troisième site le plus visité au monde
* 2ème plus grand moteur de recherche (après Google)
* En 2018, YouTube compte 37,5 millions d'utilisateurs mensuels en France
* En mars 2019, selon une étude StatistaCharts, « YouTube représenterait 37 % du trafic mondial internet sur mobile »

Utilisation croissante de la vidéo
•	Augmentation de 60% entre 2016 et 2017 du nombre d’heures
•	Plus d'un milliard d'heures de vidéo sont consommées quotidiennement par les téléspectateurs du monde entier

### Youtubeur

un YouTubeur est un Vidéaste web qui réalise et publie régulièrement du contenu vidéo sur YouTube.
Le terme peut recouvrir diverses fonctions déjà existantes telles que réalisateur, scénariste, producteur ou acteur.
Cette activité peut générer des revenus notamment grâce aux recettes de la publicité en ligne, au financement participatif, aux produits dérivés ou aux partenariats commerciaux (sponsoring, placement de produit).
