## Youtube

YouTube est un site web d’hébergement de vidéos et un média social sur lequel les utilisateurs peuvent envoyer, regarder, commenter, évaluer et partager des vidéos. 

* créé aux États-Unis en 2005
* racheté par Google en octobre 2006

* Troisième site le plus visité au monde (selon Alexa)
* 2ème plus grand moteur de recherche (après Google)

* En 2018, YouTube compte 37,5 millions d'utilisateurs mensuels en France
* En mars 2019, selon une étude StatistaCharts, « YouTube représenterait 37 % du trafic mondial internet sur mobile »

* Chaque jour, 1.5 milliard d'utilisateurs passent en moyenne 1h à voir des vidéos YouTube
* Chaque minute, 400h de vidéos sont uploadées sur YouTube
* + de la moitié des vues se font sur des mobiles


Un YouTubeur est un vidéaste web qui réalise et publie régulièrement du contenu vidéo sur YouTube.
Le terme peut recouvrir diverses fonctions déjà existantes telles que réalisateur, scénariste, producteur ou acteur.
Cette activité peut générer des revenus notamment grâce aux recettes de la publicité en ligne, au financement participatif, aux produits dérivés ou aux partenariats commerciaux (sponsoring, placement de produit).

## 4. La distribution des vues, des likes, des commentaires et l'"engagement"

L'engagement est une donnée trés importante pour les Youtubeurs car elle represente l'intérêt des clients pour la vidéo.
Youtube prend en compte cet intérêt pour :
* proposer des vidéos dans ses résultats de recherches.
* proposer des vidéos en rapport avec celles qui est en cours de visionnage

* 70 % du traffic d'une vidéo provient des résultats de recherches de Youtube


L'engagement est le rapport entre une interaction de l'utilisateur (clic sur bouton like, dislike ou la rédaction d'un commentaire) et le nombre de vue.

Plus les utilisateurs réagisse, meilleur est l'engagement et plus YouTube proposera la vidéo dans les résultats de recherche.

## Conclusion

Beaucoups de Youtubeur ne s'intéressent qu'au nombre de vue en raison du mode de rémunération proposé par YouTube de 1 dollar pour 1.000 vues (varie selon la démographie de l'audience de la chaîne, et la période dans l'année)

L'engagement est un élément déterminant pour améliorer le nombre de vues car il agit directement sur le référencement interne de YouTube.

Il y a de grandes disparités de taux d'engagement en fonction du thème (mots clés) et la catégorie des vidéos.

Il serait donc intéressant de continuer l'étude afin d'identifier plus précisément les thématiques des vidéos grâce aux outils d'analyse textuelle.
